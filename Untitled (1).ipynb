```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# ==============================================================================
# --- 1. DEFINE PARAMETERS & TRANSFER FUNCTIONS ---
# ==============================================================================

# --- Stage 1 Filter Parameters ---
C1 = 10e-9
R1 = 22e3
R2 = 62e3
R3 = 523
num1 = [-1 / (R1 * C1), 0.0]
den1 = [1.0, 2.0 / (R2 * C1), (R1 + R3) / (R1 * R2 * R3 * C1**2)]

# --- Stage 2 Filter Parameters ---
C2 = 10e-9
R4 = 18.2e3
R5 = 51.1e3
R6 = 432
num2 = [-1 / (R4 * C2), 0.0]
den2 = [1.0, 2.0 / (R5 * C2), (R4 + R6) / (R4 * R5 * R6 * C2**2)]

# --- System Combinations ---
# 1. Isolated Filter Subsystem (H1 * H2)
num_filter = np.convolve(num1, num2)
den_filter = np.convolve(den1, den2)
sys_filter = signal.TransferFunction(num_filter, den_filter)

# 2. Total AC Signal Chain (Filter * +20dB In-Amp Gain)
gain_in_amp = 3 
num_total = num_filter * gain_in_amp
sys_total = signal.TransferFunction(num_total, den_filter)

# ==============================================================================
# --- 2. CALCULATE BODE RESPONSES & BANDWIDTH ---
# ==============================================================================

# Create Custom High-Resolution Frequency Array
w_broad = np.logspace(1, 5, 2000)             
w_peak = np.linspace(6200, 6300, 5000)          
custom_frequencies = np.sort(np.concatenate((w_broad, w_peak)))

# Calculate Responses
frequencies, mag_filter, phase_filter = sys_filter.bode(w=custom_frequencies)
_, mag_total, phase_total = sys_total.bode(w=custom_frequencies)

# --- CALCULATE EXACT -3dB CUTOFFS AND BANDWIDTH ---
magnitude = mag_filter

# 1. Find the peak magnitude and its index
peak_idx = np.argmax(magnitude)
max_mag = magnitude[peak_idx]
w_peak = frequencies[peak_idx]

# 2. Define the -3dB threshold
mag_3db = max_mag - 3

# 3. Numerically interpolate to find exact frequencies where the curve crosses -3dB
# Left skirt (frequencies before the peak)
w_low = np.interp(mag_3db, magnitude[:peak_idx], frequencies[:peak_idx])
# Right skirt (frequencies after the peak - arrays must be reversed for interp to work)
w_high = np.interp(mag_3db, magnitude[peak_idx:][::-1], frequencies[peak_idx:][::-1])

# 4. Calculate Bandwidth in Hz
bw_rads = w_high - w_low
bw_hz = bw_rads / (2 * np.pi)
f_low_hz = w_low / (2 * np.pi)
f_high_hz = w_high / (2 * np.pi)
stopband_mask = frequencies > (2 * w_high)
if np.sum(stopband_mask) > 1:
    freqs_stop = frequencies[stopband_mask]
    mags_stop = magnitude[stopband_mask]
    
    # Linear fit on the log-frequency axis to get the slope (dB/decade)
    roll_off_db_per_decade = np.polyfit(np.log10(freqs_stop), mags_stop, 1)[0]
else:
    roll_off_db_per_decade = -80.0

print(f"Filter Peak: {max_mag:.2f} dB at {w_peak/(2*np.pi):.2f} Hz")
print(f"Lower Cutoff (-3dB): {f_low_hz:.2f} Hz")
print(f"Upper Cutoff (-3dB): {f_high_hz:.2f} Hz")
print(f"Bandwidth: {bw_hz:.2f} Hz")
print(f"Roll-off Rate: {roll_off_db_per_decade:.2f} dB/decade")

# ==============================================================================
# --- 3. PLOT THE GRAPH ---
# ==============================================================================

mag_3db = max_mag - 0.1

# 3. Numerically interpolate to find exact frequencies where the curve crosses -3dB
# Left skirt (frequencies before the peak)
w_low = np.interp(mag_3db, magnitude[:peak_idx], frequencies[:peak_idx])
# Right skirt (frequencies after the peak - arrays must be reversed for interp to work)
w_high = np.interp(mag_3db, magnitude[peak_idx:][::-1], frequencies[peak_idx:][::-1])

fig, ax1 = plt.subplots(figsize=(8, 3))

# --- Plot Frequency Response Curves ---
ax1.semilogx(frequencies, mag_filter, color='firebrick', linewidth=1.5, label='Filter Subsystem Only')
ax1.semilogx(frequencies, mag_total, color='blue', linewidth=2.0, label='Total System')

# --- Add Visual Markers & Sidebands ---
w_center = 2 * np.pi * 1000           
w_lower = 2 * np.pi * (1000 - 6.05)   
w_upper = 2 * np.pi * (1000 + 6.05)   

ax1.axvline(x=w_center, color='black', linestyle='--', linewidth=0.5, label='1000 Hz Carrier')
ax1.axvspan(w_lower, w_upper, color='red', alpha=0.3, label='Modulation Sidebands')
ax1.axhline(y=mag_3db, color='purple', linestyle='--', linewidth=1.5, label='-3dB Gain Threshold')

# --- Draw Cutoff Markers & Text ---
# Adjusted ymin to -60 to match the new plot boundaries
ax1.vlines(x=[w_low], ymin=-60, ymax=mag_3db, color='m', linestyle=':', linewidth=1.75)
ax1.vlines(x=[w_high], ymin=-60, ymax=mag_3db, color='c', linestyle=':', linewidth=1.75)

# Shifted text slightly down to sit cleanly below the x-axis
ax1.annotate(f'{f_low_hz:.1f} Hz', 
             xy=(w_low, -60), 
             xytext=(5, -25), textcoords='offset points', 
             ha='right', va='top', color='m', annotation_clip=False)

ax1.annotate(f'{f_high_hz:.1f} Hz', 
             xy=(w_high, -60), 
             xytext=(-3, -25), textcoords='offset points', 
             ha='left', va='top', color='c', annotation_clip=False)

# Bandwidth Text Box
ax1.annotate(f'BW = {bw_hz:.1f} Hz', 
             xy=(w_high, mag_3db), 
             xytext=(50, 10), textcoords='offset points',
             ha='left', va='center', color='white', fontweight='bold', 
             bbox=dict(facecolor='dimgray', alpha=0.8, edgecolor='blue', boxstyle='round,pad=0.5'),
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=-0.2", color='dimgray', lw=1.5))

# --- Final Formatting ---
ax1.set_ylabel('Magnitude (dB)')
ax1.grid(True, which="both", ls="-", color='0.65')
ax1.set_ylim(-60, 15) 
ax1.set_xlim([10000, 100000])

# Call the legend AFTER everything has been plotted
ax1.legend(loc='best', bbox_to_anchor=(0.02, 0.98), fontsize=10)

plt.tight_layout()
plt.savefig('bode_plot_4th_order_fixed_whole.png', dpi=500, bbox_inches='tight')
plt.show()
```
